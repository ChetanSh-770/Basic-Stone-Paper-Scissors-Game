```python
import random

# ASCII art for Rock, Paper, and Scissors
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

bang = True  # Flag to control the game loop
system = 0  # Score for the system (computer)
player = 0  # Score for the player

while bang:
    choice = input("Do you want to play Rock, Paper, Scissors? (y/n): ")
    if choice == 'y' or choice == 'Y':
        print("Enter 0 for Rock, 1 for Paper & 2 for Scissor\n")
        try:  # Handle potential errors if the user enters non-integer input
            IP = int(input("Enter your choice : "))  # Get player's input
            match = random.randint(0, 2)  # Generate computer's choice (0, 1, or 2)

            if IP == 0:  # Player chooses Rock
                print("You choose :Rock\n", rock)
                if match == 0:  # Computer chooses Rock
                    print("System Choose :Rock\n", rock)
                    print("TIE!!")
                    system += 1
                    player += 1
                elif match == 1:  # Computer chooses Paper
                    print("System Choose :Paper\n", paper)
                    print("YOU LOOSE!!")
                    system += 1
                elif match == 2:  # Computer chooses Scissors
                    print("System Choose :Scissors\n", scissors)
                    print("YOU WIN!!")
                    player += 1
            elif IP == 1:  # Player chooses Paper
                print("You choose :Paper\n", paper)
                if match == 0:  # Computer chooses Rock
                    print("System Choose :Rock\n", rock)
                    print("YOU WIN!!")
                    player += 1
                elif match == 1:  # Computer chooses Paper
                    print("System Choose :Paper\n", paper)
                    print("TIE!!")
                    system += 1  # Corrected tie logic: Both scores should increment
                    player += 1  # Corrected tie logic: Both scores should increment
                elif match == 2:  # Computer chooses Scissors
                    print("System Choose :Scissors\n", scissors)
                    print("YOU LOOSE!!")
                    system += 1
            elif IP == 2:  # Player chooses Scissors
                print("You choose :Scissors\n", scissors)
                if match == 0:  # Computer chooses Rock
                    print("System Choose :Rock\n", rock)
                    print("YOU LOOSE")
                    system += 1
                elif match == 1:  # Computer chooses Paper
                    print("System Choose :Paper\n", paper)
                    print("YOU WIN!!")
                    player += 1
                elif match == 2:  # Computer chooses Scissors
                    print("System Choose :Scissors\n", scissors)
                    print("TIE!!")
                    system += 1
                    player += 1
            else:
                print("\nINVALID INPUT!!!!!")

        except ValueError:  # Handle the error if the user input is not an integer
            print("\nInvalid input. Please enter 0, 1, or 2.")


    elif choice == 'n' or choice == 'N':
        print("Score \nSystem: {0}  Player: {1}".format(system, player))
        print("Thank You")
        break  # Exit the game loop
    else:
        print("Please enter a valid response")
```
